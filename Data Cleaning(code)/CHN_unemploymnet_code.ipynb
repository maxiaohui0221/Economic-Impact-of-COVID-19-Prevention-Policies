```python
import csv

data = []
with open('../OxCGRT_compact_subnational_v1.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in reader:
        if row[0] == 'China':
          data.append(row)
for i in range(10):
    print(data[i])
```

```python
unemployment = []
with open('China Unemployment Rate.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in reader:
        unemployment.append(row)
unemployment = unemployment[4:-1]
print(unemployment)
```

```python
interesting_columns = [8, 10, 12, 14, 16, 18, 20, 22]

def get_data(state, year):
    data2 = []
    for row in data:
        if row[2] == state and row[7].startswith(year):
            data2.append(row)
    output = []
    for column in interesting_columns:
        sum = 0
        for row in data2:
            sum += int(row[column])
        days = len(data2)
        if days == 0:
            return[0,0,0,0,0,0,0,0]
        avg = sum / days
        output.append(avg)
    return output


final_output = []

for row in unemployment:
  state = row[1]
  final_output.append([state, '2019', row[4]] + get_data(state, '2019'))
  final_output.append([state, '2020', row[3]] + get_data(state, '2020'))
  final_output.append([state, '2021', row[2]] + get_data(state, '2021'))

final_output = sorted(final_output)
print(final_output)
```

```python
with open('CHN_unemployment_combine.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    writer.writerow('State,Year,Unemployment Rate,School closing,Workplace closing,Cancel public event,Restriction on gathering,Close public transport,State at home requirement,Restrictions on internal movement,International travel control'.split(','))
    for row in final_output:
      writer.writerow(row)
```
